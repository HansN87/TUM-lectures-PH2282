# Read data from file

```python
# read data and check if format is fine
import pandas as pd
data = pd.read_csv("HeteroticOrbifoldMSSMs.csv")  # read data
data = data.drop('Unnamed: 0', axis = 1)          # drop 'Unnamed: 0' column

print("Features Y0 - Y13:\n")
print("Y0: Orbifold Geometry Label\n")
print("Y1: Number of additional U(1) gauge groups in addition to U(1)_Y")
print("Y2: Maximal rank of non-Abelian gauge groups\n\nY3: Number of exotics")
print("Y4: Number of non-Abelian- and MSSM-singlets\n")
print("Number of vectorlike MSSM-exotics:")
print("Y5: Number of additional (3,2)_1/6 + (-3,2)_-1/6")
print("Y6: Number of additional (3,1)_2/3 + (-3,1)_-2/3")
print("Y7: Number of additional (3,1)_-1/3 + (-3,1)_1/3")
print("Y8: Number of Higgs-pairs (1,2)_1/2 + (1,2)_-1/2")
print("Y9: Number of additional (1,1)_-1 + (1,1)_1\n")
print("Number of further vectorlike exotics:")
print("Y10: Number of (3,1)_x + (-3,1)_-x")
print("Y11: Number of (3,2)_x + (-3,2)_-x")
print("Y12: Number of (1,2)_x + (1,2)_-x")
print("Y13: Number of (1,1)_x + (1,1)_-x")

print("\nSize of table: ",data.shape)           # print shape of data
print("\nTable of data:")                       # show head of data
data.head()
```

# Prepare data

```python
# show different possible orbifold geometries of our dataset.
import numpy as np
print("Different labels of orbifold geometries:")
print(np.unique(data.Y0))
```

```python
# transform orbifold geometry labels to one hot encoding
data_one_hot = pd.get_dummies(data)

# check format
print("Size of table: ",data_one_hot.shape)
# show head of data
print("\nTable of data:")
data_one_hot.head()
```

```python
# transform labels of orbifold geometries to integer number
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(data.Y0)
print("Different labels of orbifold geometries: ")
print(le.classes_)
data.Y0 = le.transform(data.Y0)
# check format
print("\nSize of table: ",data.shape)
# show head of data
print("\nTable of data:")
data.head()
```

```python
# print how labels of orbifold geometries are encoded to integer
integer_mapping = {l: i for i, l in enumerate(le.classes_)}
integer_mapping
```

## Histograms

```python
import matplotlib.pyplot as plt

# plot histograms
features = list(data)                                 # store features "Yxx" of data

for i in range(data.shape[1]):
    value = np.unique(data[features[i]])              # x-values
    number = np.bincount(data[features[i]])           # number of models with x-value
    
    plt.figure(figsize = (15,10))
    plt.bar(value, number[number != 0])
    plt.title("Histogram for %s" %(features[i]))
    plt.ylabel("Number of models")
    plt.xlabel(features[i])
    plt.savefig("Plots/Histogram_of_%s" %(features[i]))
    plt.show()
```

```python
# print all values and how often they occur 

for i in range(data.shape[1]):
    print('values of feature Y%i (there are %i different values)' %(i, len(np.unique(data[features[i]]))))
    print('possible values')
    print(np.unique(data[features[i]]))
    print('number of how often each value occurs')
    print(np.bincount(data[features[i]]))
    print()
```
