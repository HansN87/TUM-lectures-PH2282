```python
from scipy.stats import norm
from scipy.stats import chi2

import numpy
import matplotlib.pyplot as plt

rvars = np.random.normal(size=int(1.e7))

# transformation x -> z: g(x)=x^2
rvars_sq = np.power(rvars, 2)

xvals = np.linspace(-4,4,1000)
yvals = norm.pdf(xvals, 0, 1)
plt.hist(rvars, bins=40, density=True)
plt.plot(xvals, yvals, "r--", label="std normal")
plt.ylabel("pdf fx(x)", fontsize=18)
plt.xlabel("x", fontsize=18)
plt.tick_params(labelsize=18)
plt.xlim([-4.0,4.0])
plt.legend(fontsize=16)
plt.show()

# as demonstrated in the lecture, the square of standard normal random variable 
# is distributed as a chisq random variable with dof = 1
# (check transformation law for random variables)

xvals = np.linspace(0, 20, 1000)
yvals = chi2.pdf(xvals, 1)
plt.hist(rvars_sq, bins=40, density=True)
plt.plot(xvals, yvals, "r--", label='chisq(dof=1)')
plt.ylabel("pdf fz(z)", fontsize=18)
plt.xlabel("z=x^2", fontsize=18)
plt.tick_params(labelsize=18)
plt.yscale('log')
plt.xlim([0.0,20.0])
plt.legend(fontsize=16)
plt.show()
```
