```python
# Scientific libraries
import numpy_html
import numpy as np

# import Pandas

import pandas as pd

# Graphic libraries

import matplotlib.pyplot as plt
%matplotlib notebook
from jupyterthemes import jtplot
plt.style.use('mike') 
jtplot.style(context='notebook', fscale=1, grid=False)
#import seaborn as sns

#optional 3ML imports

#from threeML import *
#from astromodels.xspec.xspec_settings import *
#from astromodels.xspec.factory import *

#useful libraries

from glob import glob
import copy
import collections

green = "#1DEBA6"
red = "#FF0059"
yellow = "#F6EF5B"

dark = red
mid= yellow


import stan_utility
#import warnings
#warnings.simplefilter('ignore')
```

```python
np.random.seed(1234)
```

```python
N = 4500
N_groups = 10
x1 = np.random.uniform(-1,1, size=N)

groups = np.array([int(N/N_groups) * [i] for i in range(N_groups) ]).flatten()

a = np.random.normal(loc=0,scale=3., size=N_groups)
```

```python
mu = 1.*x1 + a[groups]
```

```python

sigma = 2

y = np.random.normal(loc= mu, scale=sigma, size=N)
```

```python
fig, ax = plt.subplots()


for i in range(N_groups):
    
    idx = groups == i
    
    ax.scatter(x1[idx], y[idx])
    
    
    
    
```

```python
N_model = 100
x_model = np.linspace(-1,1,N_model)

data  = dict(N=N, N_groups=N_groups,  groups=groups+1, x1=x1, y=y, sigma=sigma, N_model=N_model, x_model=x_model)
```

```python
model = stan_utility.compile_model('hglm.stan', model_name='hglm')
```

```python
fit = model.sampling(data=data, iter=2000)
```

```python
fit.plot(pars=['beta0', 'beta1']);
```

```python
fig, ax = plt.subplots()

chain = fit.extract()
for n in range(N_groups):
    level = 95

    ax.fill_between(x_model,
                    np.percentile(chain['lines'][...,n], 50 - 0.5*level, axis=0 ),
                    np.percentile(chain['lines'][...,n], 50 + 0.5*level, axis=0 ),
                    color=dark)

    level = 68

    ax.fill_between(x_model,
                    np.percentile(chain['lines'][...,n], 50 - 0.5*level, axis=0 ),
                    np.percentile(chain['lines'][...,n], 50 + 0.5*level, axis=0 ),
                    color=mid)

#ax.plot(x_model, m_true*x_model+b_true,color='limegreen',ls='--',zorder=10000)

#ax.scatter(x_obs, y_obs, c='k')

ax.set_xlabel('x')
ax.set_xlabel('y')
```
